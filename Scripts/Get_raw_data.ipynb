```python
#Importing libraries
import os
import yfinance as yf
import pandas as pd
from fredapi import Fred
```

```python
#importing all s&p500 relating data from yfinance
sp500 = yf.download("^GSPC", start="1998-12-31", end="2026-01-01") 
sp500.to_csv("../Data/Raw_data/SP500_raw_data.csv")
```

```python
#getting all cpi data I need to calculate inflation
with open("../FRED_API_key.txt", "r") as f: #reads a txt file with my API key so it is kept private
    fred_api_key = f.read().strip()
fred = Fred(api_key = fred_api_key) #If copying code: input your FRED API into this section and delete above 2 rows
cpi_data = fred.get_series("CPIAUCSL") #gets all data on cpi for all urban consumers in the USA (seasonally adjusted)
cpi_df = pd.DataFrame(cpi_data, columns = ["CPI"])
cpi_df.index.name = "Date"
cpi_df.to_csv("../Data/Raw_data/CPI_raw_data.csv")
```

```python
#Gets data on inflation expectations from the FRED
inflation_expectations = fred.get_series("MICH") 
inflation_expectations_df = pd.DataFrame(inflation_expectations, columns = ["Inflation_expectations"])
inflation_expectations_df.index.name = "Date"
inflation_expectations_df.to_csv("../Data/Raw_data/inflation_expectations_raw_data.csv")
```
